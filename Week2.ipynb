**4. Membuat bentuk Geometri 2D dari OpenCV**

Pembuatan black image dengan tipe data int16

```python
black_img = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(black_img)
```

Kemudian menambahkan bentuk persegi panjang sesuai koordinat pt1 dan pt2

```python
cv.rectangle(black_img,pt1=(384,0),pt2=(512,150),color=(0,255,0),thickness=10)
plt.imshow(black_img)
```

Menambahkan bentuk persegi sesuai koordinat pt1 dan pt2 yang tertulis pada kode program

```python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_img)
```

Menambahkan bentuk lingkaran sesuai radius yang tertulis pada kode program

```python
cv.circle(black_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_img)
```

Menambah garis sesuai koordinat pt1 dan pt2

```python
cv.line(black_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_img)
```

Menambah text dengan font yang telah tertulis dengan ukuran yang sudah ditentukan

```python
font=cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img,text='Hello',org =(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
```

Pembuatan black image dengan tipe data int32

```python
black_img2 = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_img2)
```

inisialisasi NumPy array dengan tipe data int32

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices
```

Melakukan reshape array

```python
pts = vertices.reshape((-1,1,2)) #nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yg mewakili R,G,B dan B
pts
```

Penambahan polyline pada black image kedua

```python
cv.polylines(black_img2,[pts],isClosed=True, color=(255,0,0),thickness=5)
plt.imshow(black_img2)
```
